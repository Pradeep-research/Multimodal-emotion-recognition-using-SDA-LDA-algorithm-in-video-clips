```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
import dlib
import skvideo
from skvideo import io
from scipy.spatial import distance
import pickle
```

```python
def Calculations(video):
    for i in range(video.shape[0]):
        gray = video[i]
        detector= dlib.get_frontal_face_detector()
        predictor= dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
        faces= detector(gray)
        landmarks= []
        for face in faces:
            landmarks= predictor(gray, face)
        LM= np.zeros(2)
        for n in range(68):
            x= landmarks.part(n).x
            y= landmarks.part(n).y
            MM=np.hstack([x,y])
            LM=np.vstack([LM,MM])
        LM= np.delete(LM,0,axis=0)
        ZZ= np.flip(LM,0)
        vid_landmarks = np.zeros([68,2] ,dtype=int)
        vid_landmarks = ZZ
        dist = []
        d1 = []
        d2 = []
        d3 = []
        d4 = []
        d5 = []
        d6 = []
        d7 = []
        d8 = []
        d9 = []
        avg_dist = np.empty([10] , dtype = float)
        dst = distance.euclidean(vid_landmarks[48],vid_landmarks[36])
        d1.append(dst)
        dst = distance.euclidean(vid_landmarks[54],vid_landmarks[45])
        d2.append(dst)
        dst = distance.euclidean(vid_landmarks[37],vid_landmarks[41])
        d3.append(dst)
        dst = distance.euclidean(vid_landmarks[43],vid_landmarks[47])
        d4.append(dst)
        dst = distance.euclidean(vid_landmarks[48],vid_landmarks[54])
        d5.append(dst)
        dst = distance.euclidean(vid_landmarks[62],vid_landmarks[66])
        d6.append(dst)
        dst = distance.euclidean(vid_landmarks[33],vid_landmarks[51])
        d7.append(dst)
        dst = distance.euclidean(vid_landmarks[19],vid_landmarks[37])
        d8.append(dst)
        dst = distance.euclidean(vid_landmarks[23],vid_landmarks[43])
        d9.append(dst)
        a1 = sum(d1)
        b1 = len(d1)
        c1 = a1/b1
        avg_dist[1] = c1
        a1 = sum(d2)
        b1 = len(d2)
        c1 = a1/b1
        avg_dist[2] = c1
        a1 = sum(d3)
        b1 = len(d3)
        c1 = a1/b1
        avg_dist[3] = c1
        a1 = sum(d4)
        b1 = len(d4)
        c1 = a1/b1
        avg_dist[4] = c1
        a1 = sum(d5)
        b1 = len(d5)
        c1 = a1/b1
        avg_dist[5] = c1
        a1 = sum(d6)
        b1 = len(d6)
        c1 = a1/b1
        avg_dist[6] = c1
        a1 = sum(d7)
        b1 = len(d7)
        c1 = a1/b1
        avg_dist[7] = c1
        a1 = sum(d8)
        b1 = len(d8)
        c1 = a1/b1
        avg_dist[8] = c1
        a1 = sum(d9)
        b1 = len(d9)
        c1 = a1/b1
        avg_dist[9] = c1
        avg_dist= np.delete(avg_dist,0,axis=0)
    return avg_dist
    
```

```python
DIR = "C:\\Users\\harsh\\Desktop\\Image & Video processing\\ENTERFACE multimodal\\Database"
CATEGORIES = ["Anger","Disgust","Fear","happiness","Sadness","Suprise"]
aa = 0 # used to keep a track of number of videos which have been processed
training_data = []
for category in CATEGORIES:
    path = os.path.join(DIR,category)
    list = os.listdir(path)
    class_num = CATEGORIES.index(category)
    for i in range(190,208):
        x = os.path.join(path,list[i])
        vid = skvideo.io.vread(x)
        a = Calculations(vid)
        pp = np.hstack([a,class_num])
        training_data.append(pp)
        print(aa)
        aa = aa + 1
```

```python
pickle_out = open("Data_Video","wb")
pickle.dump(training_data , pickle_out)
pickle_out.close()
```
