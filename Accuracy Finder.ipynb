```python
import librosa
import numpy as np
import os
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectFromModel
from sklearn import svm
import sklearn.model_selection as cv
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.utils.multiclass import unique_labels
import matplotlib.pyplot as plt
%matplotlib inline
import pickle
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
pickle_in = open("Video_Data","rb")
a1 = pickle.load(pickle_in)
pickle_in.close()

pickle_in = open("Speech_Data","rb")
a2 = pickle.load(pickle_in)
pickle_in.close()

print(len(a1))
print(len(a2))
```

```python
print(a1)
```

```python
print(a2)
```

```python
X1 , Y1 = zip(*a1)
X2 , Y2 = zip(*a2)
```

```python
X = []
for i in range(0,1200):
    AA = X1[i] + X2[i]
    X.append(AA)
print(X)
```

```python
print(len(X[0]))
```

```python
print(len(X))
print(len(Y1))
```

```python
X_train, X_test, y_train, y_test = cv.train_test_split(X, Y1, test_size=0.25)
clf = LinearDiscriminantAnalysis(n_components=7)
F = clf.fit(X_train,y_train)
FF = clf.fit(X_test,y_test) 
A = clf.transform(X_train)  # X_train
B = clf.transform(X_test)  # X_test


SVM_clf = svm.SVC(kernel='linear', probability=True)

SVM_clf.fit(A,y_train)

y_pred_SVM = SVM_clf.predict(B)
print(classification_report(y_test, y_pred_SVM))

   
accuracy_SVM = accuracy_score(y_test, y_pred_SVM)
print('Accuracy of SVM:', accuracy_SVM)
```
