```python
import librosa
import numpy as np
import os
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectFromModel
from sklearn import svm
import sklearn.model_selection as cv
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.utils.multiclass import unique_labels
import matplotlib.pyplot as plt
%matplotlib inline
import pickle
```

```python
DIR = "C:\\Users\\harsh\\Desktop\\Image & Video processing\\ENTERFACE multimodal\\Database"
CATEGORIES = ["Anger Audio","Disgust Audio","Fear Audio","happiness Audio","Sadness Audio","Suprise Audio"]
mfcc_data = []
for category in CATEGORIES:
    path = os.path.join(DIR,category)
    list = os.listdir(path)
    class_num = CATEGORIES.index(category)
    for i in range(len(list)):
        x = os.path.join(path,list[i])
        A,fs= (librosa.load(x, res_type='kaiser_fast'))
        mfccs = librosa.feature.mfcc(y=A, sr=44100, n_mfcc=40)
        avg= []
        for i in mfccs:
            x1 = len(i)
            x2 = sum(i)
            x3 = x2/x1
            avg.append(x3)
        arr = avg,class_num
        mfcc_data.append(arr)
```

```python
print(len(mfcc_data))
```

```python
pickle_out = open("Speech_Data","wb")
pickle.dump(mfcc_data,pickle_out)
pickle_out.close()
```

```python
print(mfcc_data)
```

```python
X, y = zip(*mfcc_data)
```

```python
X = np.asarray(X)
y = np.asarray(y)
```

```python
X_train, X_test, y_train, y_test = cv.train_test_split(X, y, test_size=0.25)
SVM_clf = svm.SVC(kernel='linear', probability=True)


SVM_clf.fit(X_train, y_train)

y_pred_SVM = SVM_clf.predict(X_test)
print(classification_report(y_test, y_pred_SVM))

   
accuracy_SVM = accuracy_score(y_test, y_pred_SVM)
print('Accuracy of SVM:', accuracy_SVM)
```

```python
from sklearn.tree import DecisionTreeClassifier
X_train, X_test, y_train, y_test = cv.train_test_split(X, y, test_size=0.25)
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
predictions = dtree.predict(X_test)
print(classification_report(y_test, predictions))

   
accuracy_SVM = accuracy_score(y_test, predictions)
print('Accuracy of SVM:', accuracy_SVM)
```
